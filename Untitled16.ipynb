```python
#minutes column

uber['Minutes'] = uber['END_DATE']-uber['START_DATE']
uber.head()
```

```python
plt.figure(figsize=(10,8))
sns.lineplot(data = uber,
             x = uber['Minutes'].dt.total_seconds() / 60,
             y = 'MILES',
             palette='Set1')
plt.xlabel('Minutes')
plt.ylabel('Miles')
plt.title('Miles vs Minutes')
plt.show()
```

```python
def round(x):
  if x['START'] == x['STOP']:
    return 'yes'
  else:
    return 'no'

uber['Round_Trip'] = uber.apply(round, axis = 1)

uber.head()
```

```python
uber.drop(columns = ['Months'], inplace = True)
```

```python
uber['Months'] = pd.DatetimeIndex(uber['START_DATE']).month


dict = {1:'Jan', 2:'Feb', 3:'Mar', 4:'Apr', 5:'May', 6:'Jun', 7:'Jul', 8:'Aug', 9:'Sep', 10:'Oct', 11:'Nov', 12:'Dec'}

uber['Months'] = uber['Months'].map(dict)
```

```python
uber.head()
```

```python
# prompt: generate a code to create WEEKDAYS column with weekdays [ 'monday,tuesday,wednessday,thursday, friday, saturday, sunday]

uber['WEEKDAYS'] = uber['START_DATE'].dt.day_name()

uber.drop(columns= ['Weekdays'], inplace= True)
uber.head()
```

##  what is the longest and shortest ride duration for each category


```python
uber['Ride_duration'] = (uber['END_DATE'] - uber['START_DATE']).dt.total_seconds() / 60
uber.drop(columns = ['duration'],inplace = True)
uber.head()
```

```python
duration = uber.groupby(['CATEGORY'])['Ride_duration'].agg(['min','max'])
duration
```

## **what is the total number of Miles travel per month and which month had the most miles travelled**

```python
total_no_of_miles = uber.groupby(['Months'])['MILES'].sum().head(5)
print(total_no_of_miles)

plt.figure(figsize=(10,8))
total_no_of_miles.plot(kind= 'bar',
                       color = ['#FFD586','#FFAAAA','#FFE99A'])

plt.title("TOTAL NUMBER OF MILES TRAVELLED PER MONTH")
plt.show()
```

## **what is the average number of rides per day for each day of the week**

```python
week = uber.groupby(['WEEKDAYS'])['MILES'].sum()
avg = week / uber['WEEKDAYS'].nunique()
print(avg)

plt.figure(figsize=(10,8))
avg.plot(kind= 'bar',
        color = ['#AA60C8','#D69ADE','#EABDE6','#FFDFEF'])

plt.title("AVERAGE NUMBER OF RIDES PER DAYS FOR EACH DAYS IN WEEK")
plt.show()
```
